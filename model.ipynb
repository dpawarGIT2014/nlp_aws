### Feature Engineering with nltk

```python
import multiprocessing
from multiprocessing import Pool
import csv
import nltk
nltk.download('punkt')
```

```python
df['Review_tokenize'] = df['Review'].str.lower()

df['Review_tokenize'] = df.apply(lambda row: nltk.word_tokenize(row['Review_tokenize']), axis=1)

df.head()
```
